```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
rawDataset = pd.read_csv('heart_attack.csv')

rawDataset
```

```python
dataset = rawDataset[['age', 'trestbps', 'heart_disease']].copy()
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

dataset
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size = 0.25,
                                                   random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1)
                     , y_test.reshape(len(y_test), 1))
                    ,1))
```

```python
# confusion matrix: Age & Resting BP to Heart Attack
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# accuracy score: Age & Resting BP to Heart Attack
accuracy_score(y_test, y_pred)
```

```python
from matplotlib.colors import ListedColormap

X_set, y_set = sc.inverse_transform(X_train), y_train

X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10
                              , stop = X_set[:, 0].max() + 10
                              , step = 0.25)
                    , np.arange(start = X_set[:, 1]. min() - 1000
                               , stop = X_set[:, 1].max() + 1000
                               , step = 0.25))

plt.contourf(X1
            , X2
            , classifier.predict(sc.transform(np.array([X1.ravel()
                                                       , X2.ravel()]).T)).reshape(X1.shape)
            , alpha = 0.75
            , cmap = ListedColormap(('salmon', 'dodgerblue')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0]
               , X_set[y_set == j, 1]
               , color = ListedColormap(('red', 'blue'))(i)
               , label = j)
    
plt.title('Age & Resting BP to Heart Attack (Training set)')
plt.xlabel('Age')

plt.ylabel('Trestbps')

plt.legend()
```

```python
from matplotlib.colors import ListedColormap

X_set, y_set = sc.inverse_transform(X_test), y_test

X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10
                              , stop = X_set[:, 0].max() + 10
                              , step = 0.25)
                    , np.arange(start = X_set[:, 1]. min() - 1000
                               , stop = X_set[:, 1].max() + 1000
                               , step = 0.25))

plt.contourf(X1
            , X2
            , classifier.predict(sc.transform(np.array([X1.ravel()
                                                       , X2.ravel()]).T)).reshape(X1.shape)
            , alpha = 0.75
            , cmap = ListedColormap(('salmon', 'dodgerblue')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0]
               , X_set[y_set == j, 1]
               , color = ListedColormap(('red', 'blue'))(i)
               , label = j)
    
plt.title('Age & Resting BP to Heart Attack (Test set)')
plt.xlabel('Age')

plt.ylabel('Trestbps')

plt.legend()
plt.show()
```

```python
# Equation model: Age & Resting BP to Heart Attack
model = LogisticRegression()
model.fit(X, y)

print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)

print('Logit(p) = ', model.intercept_[0], '+', model.coef_[0][0], '* age +', model.coef_[0][1], '* Trestbps')
```

```python
## Logit(p) =  -4.1941884865940615 + 0.04666707970336519 * age + 0.01125905493880517 * Trestbps

def heartAttackProbability(Age, Trestbps):
    p = -4.1941884865940615 + 0.04666707970336519 * Age + 0.01125905493880517 * Trestbps
    print("Probability: ", p)

heartAttackProbability(67, 130)
```

```python


def heartAttackProbabilityv2(Age, Trestbps):
    log_odds = -4.1941884865940615 + 0.04666707970336519 * Age + 0.01125905493880517 * Trestbps
    odds = np.exp(log_odds)
    p = odds / (1 + odds)
    print("Probability: ", p)

heartAttackProbabilityv2(97, 220)
```

```python
def heartAttackChance(Age, Trestbps):
    e = 2.718281828459045
    proba = 1 / (1 + e**(-(-4.1941884865940615 + 0.04666707970336519 * Age + 0.01125905493880517 * Trestbps)))
    print('Heart attack probability: ', proba)
    
heartAttackChance(97, 220)
```
