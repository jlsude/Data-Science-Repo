```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
dataset = pd.read_csv("wine-dataset.csv")
print("dataset: {}".format(dataset.shape))
print("Columns: {}".format(dataset.columns))
z = dataset.head(5)
z
```

```python
#Relationship of points to price per province
sns.scatterplot(data = z, x = "price", y = "points",  hue = "province")
print('Relationship of points to price per province (5 province)')
```

```python
#relationship of variety to price
sns.boxplot(data = z, x = 'price', y = 'variety')
print('Relationship of the Variety to Price (5 varieties)')
```

```python
#Count of variety per country
sns.countplot(x=z["country"])
print('Count of variety per country (5 Varieties)')
```

```python
#The Median price per country
a = dataset.head(26)
g = a.groupby(['country'])['price'].median().reset_index()
sns.boxplot( x = 'country', y = 'price', data = g)
print('The Median price per country')
```

```python
#Features of heatmap (All features)
sns.heatmap(z.corr())
```
