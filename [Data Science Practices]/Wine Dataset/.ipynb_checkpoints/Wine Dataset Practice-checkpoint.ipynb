```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rawdataset = pd.read_csv('winemag-data-130k-v2.csv')

rawdataset.head()
```

```python
rawdataset.info()
```

```python
rawdataset.describe()
```

```python
rawdataset.isnull().sum()
```

```python
sns.barplot(x=rawdataset['country'].value_counts().head(10).values, y=rawdataset['country'].value_counts().head(10).index)
plt.title('Top Wine Maker Countries')

```

```python
WinevsPriceCountries = rawdataset[['country','points', 'price']].head(100)
Top100WinevsPriceCountries = WinevsPriceCountries.sort_values(by=['points'], ascending=False)
Top100WinevsPriceCountries

sns.scatterplot(data=Top100WinevsPriceCountries, 
                x=Top100WinevsPriceCountries['points'], 
                y=Top100WinevsPriceCountries['price'],
               hue=Top100WinevsPriceCountries['country'])
plt.title('Top 100 Wine vs Price and their origin')
```

```python
datapie = WinevsPriceCountries['country'].value_counts()

plt.pie(datapie, labels=datapie.index, autopct='%1.1f%%')
plt.title('Countries at Top 100 Wine')
```

```python
#Average score given by each taster
sorted_result = rawdataset.groupby('taster_name')['points'].mean().sort_values(ascending=False)

# Create the barplot
ax = sns.barplot(x=sorted_result.values, y=sorted_result.index, palette='Set2')

# Annotate the bars with their respective values
for i, v in enumerate(sorted_result.values):
    ax.text(v + -8, i, str(round(v, 2)), va='center', fontsize=10, color='black')

# Set labels and title
plt.xlabel('Average Points')
plt.ylabel('Taster Name')
plt.title('Average Points by Taster Name')

# Show the plot
plt.show()





```
